```python
sys_bp_max = month2_df['SystolicBP'].max()
sys_bp_min = month2_df['SystolicBP'].min()

dia_bp_max = month2_df['DiastolicBP'].max()
dia_bp_min = month2_df['DiastolicBP'].min()
```

```python
# function to regulate systolic blood pressure
def regulate_sys_bp(sys_bp):
    if sys_bp > sys_bp_max:
        return sys_bp_max
    elif sys_bp < sys_bp_min:
        return sys_bp_min
    else:
        return sys_bp
    
# function to regulate diastolic blood pressure
def regulate_dia_bp(dia_bp):
    if dia_bp > dia_bp_max:
        return dia_bp_max
    elif dia_bp < dia_bp_min:
        return dia_bp_min
    else:
        return dia_bp
```

```python
month2_df["SystolicBP2"] = month2_df["SystolicBP2"].apply(regulate_sys_bp)
month2_df["DiastolicBP2"] = month2_df["DiastolicBP2"].apply(regulate_dia_bp)
```
